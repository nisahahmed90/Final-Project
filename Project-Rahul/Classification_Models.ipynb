```python
# Import Dependancies
import pandas as pd
from path import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Load data
file_path = "../diabetes-filtered.csv"
df2 = pd.read_csv(file_path)
df2
```

```python
#Encode Reamditted Column

df2['readmitted']=df2['readmitted'].replace('NO', 0)
df2['readmitted']=df2['readmitted'].replace('>30', 1)
df2['readmitted']=df2['readmitted'].replace('<30', 2)
```

```python
#Encoding the Data
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse = False)

#Encode Column Race
encode_df = pd.DataFrame(enc.fit_transform(df2.race.values.reshape(-1,1)))

#Rename encoded columns
encode_df.columns = enc.get_feature_names(['race'])
encode_df.head()

# Merge Dataframes
df2 = df2.merge(encode_df,left_index=True,right_index=True)

#Encoding the Data
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse = False)

#Encode Column Race
encode_df2 = pd.DataFrame(enc.fit_transform(df2.gender.values.reshape(-1,1)))

#Rename encoded columns
encode_df2.columns = enc.get_feature_names(['gender'])

# Merge Dataframes
df2 = df2.merge(encode_df2,left_index=True,right_index=True)
```

```python
# Drop Encoded Columns and ID Column

df2.drop(columns = ['race','gender','id'], inplace = True)
```

```python
# Classify X and y Data

y = df2['readmitted']
X = df2.drop(columns = ['readmitted'])
```

```python
# Create testing and training set

X_train,X_test,y_train,y_test = train_test_split(X,y, train_size = 0.8, random_state = 6, stratify = y)
```

```python
import sklearn as skl
# Create scaler instance
X_scaler = skl.preprocessing.StandardScaler()

# Fit the scaler
X_scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Resample the training data with the RandomOversampler

from collections import Counter
Counter(y_train)
from imblearn.over_sampling import RandomOverSampler
sampler = RandomOverSampler(random_state =6)
X_resampled, y_resampled = sampler.fit_resample(X_train_scaled, y_train)

Counter(y_resampled)
```

Randon Forest Classifier

```python
# Create a random forest classifier.
model = RandomForestClassifier(n_estimators=128, random_state=6) 
# Fitting the model.
model = model.fit(X_resampled, y_resampled)
```

```python
# Making predictions using the testing data.
y_pred = model.predict(X_test)
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, y_pred)
acc_score
```

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual NO", "Actual >30",'Actual <30'], columns=["Predicted NO", "Predicted >30",'Predicted <30'])

cm_df
```

```python
print(classification_report(y_test, y_pred))
```

EasyEnsemble Classifier

```python
# Train the EasyEnsembleClassifier
from imblearn.ensemble import EasyEnsembleClassifier
model1 = EasyEnsembleClassifier(random_state= 6)

model1.fit(X_resampled,y_resampled)
```

```python
#Obtaining predicted values
y_pred1 = model1.predict(X_test)

# Calculated the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(y_test, y_pred1)
```

```python
# Print the imbalanced classification report

from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, y_pred1))
```

ANN Model

```python
#Import Tensorflow module
import tensorflow as tf

# Create the Keras Sequential model
nn_model = tf.keras.models.Sequential()
```

```python
# Add our first Dense layer, including the input layer
inputs = len(X_resampled[0])

nodes_first_layer = 30
nodes_second_layer = 15
nodes_third_layer = 7

# Input Layer
nn_model.add(tf.keras.layers.Dense(units=nodes_first_layer, activation="tanh", input_dim=inputs))

# Second Layer
nn_model.add(tf.keras.layers.Dense(units=nodes_second_layer, activation="tanh"))

# Third Layer
nn_model.add(tf.keras.layers.Dense(units=nodes_third_layer, activation="tanh"))

# Output layer
nn_model.add(tf.keras.layers.Dense(units = 1, activation="sigmoid"))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn_model.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
